```python
#Importing necessary libraries
import pandas as pd
import timeit
from itertools import combinations 
import multiprocessing as mp
import time
import numpy as np
import math
```

```python
#Read the country - ASN matrix file
filename = "final-v2"
df_fin = pd.read_csv("Data/WrittenData/output-df-" + filename + ".txt", sep=",", header=None)
df_fin.columns = ["ASN","to_IP","latency", "Country"]
```

```python
df_final = df_fin.pivot_table('latency', 'Country', 'ASN')
```

```python
df = pd.DataFrame(df_final)
print(df.min())
```

```python
df
```

```python
#Obtain combinations of 4 for all ASN
nASN = df.shape[1]
c = [x for x in combinations(list(range(nASN)), 4) ]
c = list(map(lambda elem: list(elem) , c))
```

```python
#Getting the sum of min values of latency for selected ASN 
df.iloc[:, [0, 1, 2, 3]].min(1).sum()
```

```python
df2 = df.values
df2
```

```python
#Exhaustive enumeration algortihm
def enumerationAlg(listIndex):
    i = listIndex[0]
    j = listIndex[1]
    k = listIndex[2]
    l = listIndex[3]
    return (np.nansum(np.nanmin(df2[:, [i, j, k, l]], axis = 1))) 
```

```python
len(c)
```

```python
import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer / denom
```

```python
ncr(163,4)
```

#The number of combinations is same above. Hence, the number of combinations is optimal. 
Do not run below.

```python
#Running the algorithm for the combinations
st = time.time()
results = list(map(enumerationAlg , c))
et = time.time()
print(round(et-st,2))
```

### Calculating the same with NAs filled with mean values ###

```python
#Convert NA to mean to obtain the average values for all countries 
df_final2 = df_final.apply(lambda x: x.fillna(x.mean())) 
df_final2.head()
```

```python
df_mean = pd.DataFrame(df_final2)
df_mean2 = df_mean.values 

#Exhaustive enumeration algortihm
def enumerationAlg2(listIndex):
    i = listIndex[0]
    j = listIndex[1]
    k = listIndex[2]
    l = listIndex[3]
    return (np.sum(np.min(df_mean2[:, [i, j, k, l]], axis = 1))) 
```

```python
#Running the algorithm for the combinations
st = time.time()
results = list(map(enumerationAlg2 , c))
et = time.time()
print(round(et-st,2))
```
