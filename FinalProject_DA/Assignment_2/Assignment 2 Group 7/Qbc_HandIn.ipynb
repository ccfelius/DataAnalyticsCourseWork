#### Reading from RIPE db ####

```python
# Probe id is needed ("prb_id")
# Ip adress destination("dest_addr")
# Note that RIPE dataset also supports IPv6 adresses
# Filter only IPv4
# but first check if dst_addr exists in entry

dst_addr = []
probe_id = []
avg = []
ASN = []

filenr = 17

filename = "Data/RIPE/ping-2020-02-20T" + str(filenr).zfill(2) + "00.bz2"

begin = time.time()

with bz2.BZ2File(filename, "rb") as file:
    for line in file:
        jsonline = json.loads(line)
        if 'dst_addr' in jsonline and 'avg' in jsonline and 'prb_id' in jsonline and ':' not in jsonline["dst_addr"] and jsonline["prb_id"] in probs["prb_id"].values and jsonline["avg"] > 0:
            probe_id.append(jsonline["prb_id"])
            avg.append(jsonline["avg"])
            dst_addr.append(jsonline["dst_addr"])
            
end = time.time()

print("Took " + str((end-begin)/60) + " minutes")
```

#### Binary Search Algorithm Complexity ####

The Binary Search Algorithm has a complexity of O(logn). This means, since the IP dataset has **174972** entries, there are at maximum log(174972)/log(2) = **17.42** iterations needed to find the range of the decimal IP-number. The number of iterations needed must be therefore lower then or equal to 17.

 The complexity is explained in another file as requested per feedback 

**'IP_list'** is now the list that contains the ip and country pair. We will use this list to match the country of the IP-adress into the other DF **'ASN_IP'**

```python
import time

## Check whether dest_addr occurs in IP dataset
# First convert to decimal IP
# Then use binary search to look up the range
# Then Check if range is from a European country
# Store only IP'adresses from European Country's

ipList = []

### **** Make sure this is the right filenr *** ###

start2 = time.time()

log = 0
for i in ASN_IP["to_IP"]:
    log += 1
    index = binarySearch(int(ipaddress.ip_address(i)), list(IPdata["startIP"]))[0]
    if IPdata.iloc[index]['cc'] in countryCodesEU:
        ipList.append([i, IPdata.iloc[index]['Country']])

                
end2 = time.time()
dur = (end2-start2)/60
print("Running took: " + str(dur) + " Minutes")
```
